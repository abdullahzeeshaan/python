```python
F_Num = float(input("Enter your first number: "))
S_Num = float(input("Enter your second number: "))
Operator = input("Choose an operator('+', '-', '*', '/'):")
Name = input("Enter your name: ")

if Operator == "+":
  Answer = F_Num + S_Num

if Operator == "-":
  Answer = F_Num - S_Num

if Operator == "*":
  Answer = F_Num * S_Num

if Operator == "/":
  Answer = F_Num / S_Num

else:
  print("Invalid Operator")

print(f'''{F_Num} {Operator} {S_Num} = {Answer}''')

if F_Num > 100:
  print(f'''Hi {Name}!''')
```
