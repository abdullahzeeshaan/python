# **Reversed Words In Sentence**

```python
sentence = input("Enter the sentence: ")

sentence = sentence.split()
print(sentence)

reversed_words = sentence[::-1]
print(reversed_words)

reversed_sentence = ' '.join(reversed_words)

print(reversed_sentence)
```

# **Check Palindrome**

```python
word = input("Enter the word: ")

if word == word[::-1]:
  print('This word is palindrome')

else:
  print('This word is not palindrome')
```

# **Cunt Vowels And Consonants**

```python
sentence = input("Enter the sentence: ").lower()
vowels = 'aeiou'

vowels_count = sum(1 for char in sentence if char in vowels)
consonants_count = sum(1 for char in sentence if char.isalpha() and char not in vowels)

print(f'Vowels {vowels_count}, Consonants {consonants_count}')
```

# **Sum Of Digits**

```python
num = input("Enter the number: ")
digits_sum = sum(int(digit) for digit in num if digit.isdigit())

print(f'sum of the digits: {digits_sum}' )
```
