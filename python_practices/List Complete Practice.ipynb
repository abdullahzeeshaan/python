#  **1. Create List**

```python
# Homogenous List

List = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(List)
```

```python
# Heterogenous List

List1 = [1,3,7, 'hello', 5+2j, False]
print(List1)
```

```python
# Multidimensional List
# 2D List

List2 = [[1,2,3,4,5,6]]
print(List2)
```

```python
# Multidimensional List
# 3D List

List3 = [[[1,2],[3,4],[5,6]]]
print(List3)
```

```python
students = ['abdullah', 'abdul rehman', 'abdul subhan', 'aleem', 'adeem', 'arif', 'razzaq', 'rizwan', 'haseeb', 'adan']
print(students)
```

```python
students
```

```python
students[1], students[3], students[5], students[7], students[9]
```

```python
# Method of update a list

students[2] = "ahmed"
print(students)
```

```python
# method of delete a list value

del students[3]
students
```

```python
# List in list updation

subjects = [students, 'maths', 'chemistry', 'physics', 'urdu']
print(subjects)
```

```python
# Adding a new value in list

subjects.append('computer')
print(subjects)
```

```python
# Insert use for the value in specific location in a list

subjects.insert(2, 'Islamiat')
print(subjects)
```

```python
# We also initiaize the list with another variable

course = subjects
print(course)
```

```python
# Remove a list value

abc = [['abdullah', 'abdul rehman', 'abdul subhan'], 'maths', 'Islamiat', 'chemistry']
abc.remove('chemistry')
print(abc)
```

```python
# Use to extend values by adding new values

abc.extend(['physics', 'computer'])
print(abc)
```

# **List Indexing & Slicing**

```python
# Positive Indexing & Slicing

num = [1, 2, 3, 4, 5, 6, 7, 8, 9]
num[7]
```

```python
num[3:8]
```

```python
num[5:]
```

```python
num[:6]
```

```python
num[0:9:2]
```

```python
# Negative Indexing & Slicing

number = [10, 20, 30, 40, 50, 60, 70, 80, 90]
number[-7]
```

```python
number[-7:-3]
```

```python
number[-5:]
```

```python
number[:-3]
```

```python
number[-8:-2:2]
```

```python
number[::-1]
```
