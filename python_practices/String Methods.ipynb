#**1. LOWER FUNCTION**

```python
String = "ATGFFHGhghtdhAhAhtAUvATUbAGV"

Small = String.lower()
print(Small)
```

#**2. UPPER FUNCTION**

```python
String = "ATGFFHGhghtdhAhAhtAUvATUbAGV"

Small = String.upper()
print(Small)
```

# **3. STRIP FUNCTION**

```python
string = "                      A                  hgDSfS   gcH GFGC ffg GCg                    "
var = string.strip()
print(var)
```

#  **4. REPLACE FUNCTION**

```python
abd = "Apple "
dfg = abd.replace('Apple','Mango')
print(dfg)
```

# **5. SPLIT FUNCTION**

```python
QWE = "GHGggnbGCNBBGJH  HGHGgfvhgHGVHG GHGghVgf%&^*^&"
RTY = QWE.split()
print(RTY)
```

# **6. JOIN FUNCTION**

```python
IOP = "             ".join(RTY)
print(IOP)
```

# **7. FIND FUNCTION**

```python
ZXC = "DFFF   jghc  HGFC H n  fHvj"
VBN = ZXC.find("n")
print(VBN)
```

# **8. STARTSWITH FUNCTION**

```python
Name = "My name is Abdullah"
Mane = Name.startswith("my")
print(Mane)
```

# **9. ENDSWITH FUNCTION**

```python
Nam = "My name is Abdullah"
Man = Nam.endswith("Abdullah")
print(Man)
```

#  **10. IS DIGIT FUNCTION**

```python
num = "127627"
Is_num = num.isdigit()
print(Is_num)
```

# **11. TITLE FUNCTION**

```python
tit = "HGHG hgv hjg hjb"
le = tit.title()
print(le)
```

# **12. CAPITALIZE FUNCTION**

```python
cap = "pagal hfhg"
pital = cap.capitalize()
print(pital)
```

# **13. ISALPHA FUNCTION**

```python
alpha = ("jhgvhgvjfgfejhjhgjhhv")
alp = alpha.isalpha()
print(alp)
```

# **14. SWAPCASE FUNCTION**

```python
swap = "FDXGFjhjh hvhnvnHGVH jhgjHVnv HVnb"
cas = swap.swapcase()
print(cas)
```

#  **15. ISALNUM FUNCTION**

```python
al = "HGCGF34567hghghgfhGvhgFhgHF6hgft5ttuI"
tyu = al.isalnum()
print(tyu)
```
