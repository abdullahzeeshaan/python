#  **Lists Function**

```python
# Append Function

a = [1,2,3]
a.append(4)
print(a)
```

```python
# Extend Function

e = [1,2,3,4]
e.extend([5])
print(e)
```

```python
# Insert Function

i = [1,2,4,5]
i.insert(0,3)
print(i)
```

```python
# Remove Function

r = [1,2,3,4,5]
r.remove(5)
print(r)
```

```python
# Pop Function

p = [1,2,3,4]
p.pop()
print(p)
```

```python
# Clear Function

c = [1,2,3,4,5]
c.clear()
print(c)
```

```python
# Index Function

ind = [10,20,30,40,50]
value = ind.index(50)
print(value)
```

```python
# Count Function

co = [1,2,3,4,5,4,3,2,1]
Value = co.count(6)
print(Value)
```

```python
x = [1,2,3,4,5]
y = x.copy()
print(y)

z = y.extend(y)
print(z)
```
