# **Finding Factorial Using While Loop**

```python
num = int(input("Enter the number: "))

factorial = 1

if factorial > 0:
  while (num > 0):
    factorial = factorial * num
    num = num - 1

  print(factorial)

else:
  print('Please enter the positive number')
```

# **Guessing Game**

```python
import random

right_num = random.randint(1, 100)
count = 1

guess = int(input("Enter the number between 1-100: "))

while (right_num != guess):
  if right_num > guess:
    print('Guess Higher')

  else:
    print('Guess Lower')

  guess = int(input("Enetr the number between 1-100: "))
  count += 1

print('Correct Answer')
print(f'''You took a {count} attempts''')
```
