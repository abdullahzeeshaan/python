# **Guessing game**

```python
import random

jackpot = random.randint(1, 100)
count = 1

guess = int(input("Enter the number: "))

while (guess != jackpot):
  if guess < jackpot:
    print("Guess Higher, again guess")

  else:
    print("Guess Lower, again guess")

  guess = int(input("Enter the number: "))
  count += 1

print("Correct Answer")
print("You took", count, "attempts")
```

# **Find the factorial of the given number**

```python
num = int(input("Enter the number: "))

factorial = 1

if num > 0:
  while (num > 0):
    factorial = factorial * num
    num = num - 1

  print(factorial)

else:
  print("Please input the positive numbers")
```

# **Sum of N Natural numbers**

```python
num = int(input("Enter the number: "))

count = 0
i = 1

while (i <= num):
  count = count + i
  i = i + 1

print(count)
```
