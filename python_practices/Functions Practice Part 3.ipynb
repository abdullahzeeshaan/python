# **Returning A Dictionary With Function**

```python
def build_person(first_name, last_name, age = ''):
  person = {'First_name': first_name, 'Last_name': last_name}

  if age:
    person['age'] = age
    return person

build_person('Abdullah', 'Zeeshan', '20')
```

```python
build = build_person('Abdullah', 'Zeeshan', '18')
```

```python
def pizza(*toppings):
  for i in toppings:
    print(i)

pizza('Papereno', 'Double cheese', 'Chicken tikka')
```

```python
def build_profile(first, last, **user_profile):
  profile = {}
  profile['First name'] = first
  profile['Last name'] = last

  for key, value in user_profile.items():
    profile[key] = value

  return profile

build_profile('Abdulllah', 'Zeeshan')
```

```python
build_profile('Abdullah', 'Zeeshan', Education = 'Intermediate', College = 'Cantt College')
```

```python
def avg(arr):
  total = 0
  for i in arr:
    total = total + i

  return total/len(arr)

avg([1,2,3,4,5])
```

# **Calculator**

```python
def calculator(arr, task = ''):

# Addition
  if task == 'add':
    total = 0
    for add in arr:
      total = total + add
    return total

# Subtraction
  elif task == 'sub':
    total = 0
    for sub in arr:
      total = sub - total
    return total

# Multiplication
  elif task == 'mul':
    total = 1
    for mul in arr:
      total = total * mul
    return total

# Division
  elif task == 'div':
    total = 1
    for div in arr:
      total = div / total
    return total

# Average
  elif task == 'avg':
    total = 0
    for avg in arr:
      total =total + avg
    return total / len(arr)

# Square
  elif task == 'sqr':
    power = int(input("Enter the power"))
    return[sqr**power for sqr in arr]

# Even List
  elif task == 'even':
    even_list = []
    for even in arr:
      if even % 2 == 0:
        even_list.append(even)
    return even_list

# odd List
  elif task == 'odd':
    odd_list = []
    for odd in arr:
      if odd % 2 == 1:
        odd_list.append(odd)
    return odd_list

  else:
    print(arr)

calculator([2,3,4], 'add')
```
