# **Factorial of the number**

```python
num = int(input("Enter the number: "))

factorial = 1

if num < 0:
  print("Factorial does not exist for the negitive numbers.")

elif num == 0:
  print("The factorial of the 0 is 1.")

else:
  for i in range(1, num + 1):
    factorial *= i

  print("The factorial of the", num, "is:", factorial)
```
