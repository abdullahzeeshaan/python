```python
# Problem 4
# Count Alphabets, Digits, Spaces and Special Charcters

fh = open(r'/content/c.txt', 'r')
data = fh.read()

a, d, s, sc = 0, 0, 0, 0

for char in data:
  if char.isalpha():
    a += 1

  elif char.isdigit():
    d += 1

  elif char.isspace():
    s += 1

  else:
    sc += 1

print('Alphabets: ', a)
print('Digits: ', d)
print('Spaces: ', s)
print('Special Characters: ', sc)
```

```python
# Problem 5
# Count Alphabets, Digits, Spaces and Special Charcters in each line of file

fh = open(r'/content/c.txt', 'r')
ls = fh.readlines()

for ln in ls:
  a, d, s, sc = 0, 0, 0, 0
  for char in ln:
    if char.isalpha():
      a += 1

    elif char.isdigit():
      d += 1

    elif char.isspace():
      s += 1

    else:
      sc += 1

  print(ln)
  print('Alphabets: ', a)
  print('Digits: ', d)
  print('Spaces: ', s)
  print('Special Characters: ', sc)
```
