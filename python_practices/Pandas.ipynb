```python
df['Total'] = df['Math'] + df['Physics'] + df['Chemistry'] + df['English']
df
```

```python
df['Percentage'] = df['Total'] / 400 * 100
df
```

```python
def assignGrade(per):
  grade = None

  if per >= 80:
    grade = 'A+'

  elif per >= 70:
    grade = 'A'

  elif per >= 60:
    grade = 'B'

  elif per >= 50:
    grade = 'C'

  else:
    grade = 'F'

  return grade

df['Grade'] = [assignGrade(per) for per in df['Percentage']]
df
```

```python
df[df['Percentage']>75][['Name', 'Roll No', 'Grade']]
```

```python
df['Percentage'].max()
```

```python
df['Percentage'].min()
```

```python
df[df['Percentage'] == df['Percentage'].max()][['Name', 'Roll No','Grade']]
```

```python
df[df['Percentage'] == df['Percentage'].min()][['Name', 'Roll No', 'Grade']]
```
