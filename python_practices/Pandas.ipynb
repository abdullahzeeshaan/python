```python
result = {'Name': ['Abdullah', 'Abdul Rehman', 'Abdul Subhan', 'Nasir', 'Ahmed'],
          'Roll No': [1010, 1020, 1030, 1040, 1050],
          'Math': [84, 88, 92, 76, 62],
          'Chemistry': [79, 70, 67, 69, 56],
          'Physics': [89, 76, 72, 71, 66],
          'English': [69, 71, 65, 62, 59]}

df = pd.DataFrame(result)
df
```

```python
df['Total'] = df['Math'] + df['Physics'] + df['Chemistry'] + df['English']
df
```

```python
df['Percentage'] = df['Total'] / 400 * 100
df
```

```python
def assignGrade(per):
  grade = None

  if per >= 80:
    grade = 'A+'

  elif per >= 70:
    grade = 'A'

  elif per >= 60:
    grade = 'B'

  elif per >= 50:
    grade = 'C'

  else:
    grade = 'F'

  return grade

df['Grade'] = [assignGrade(per) for per in df['Percentage']]
df
```

```python
df[df['Percentage']>75][['Name', 'Roll No', 'Grade']]
```

```python
df['Percentage'].max()
```

```python
df['Percentage'].min()
```

```python
df[df['Percentage'] == df['Percentage'].max()][['Name', 'Roll No','Grade']]
```

```python
df[df['Percentage'] == df['Percentage'].min()][['Name', 'Roll No', 'Grade']]
```
