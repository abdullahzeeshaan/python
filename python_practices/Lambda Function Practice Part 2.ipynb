# **Basic Function Syntax**

```python
def square(number):
  return number ** power

num = int(input("Enter the number: "))
power = int(input("Enter the power: "))

result = square(num)
print(result)
```

# **Function With Multiple Parameters**

```python
def add(x, y):
  return x + y

add(4, 6)
```

# **Polymorphism Function**

```python
def multiply(x, y):
  return x * y

print(multiply(5, 4))
print(multiply('T', 4))
print(multiply(5, 'U'))
```

# **Function Returning Multiple Values**

```python
import math

def circle_status(radius):
  area = math.pi * radius ** 2
  circumpherence = 2 * math.pi * radius
  return area, circumpherence

a, c = circle_status(4)
print(a)
print(c)
```
