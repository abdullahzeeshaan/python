```python
F_Name = input("Enter the first name: ")
L_Name = input("Enter the last name: ")
Y_O_B = int(input("Enter your year of birth: "))

Full_Name = F_Name + " " + L_Name
Full_Name = Full_Name.title()
print(Full_Name)

Age = 2024 - Y_O_B
print(Age)

if Age >= 18:
  print(f'''Hi {Full_Name}! You are {Age} years old and you are adult.''')

else:
  print(f'''Hi {Full_Name}! You are {Age} years old and you are teenager.''')
```
