```python
List = list(input("Enter the list: "))

def get_min_max(number):
  return min(number), max(number)

min_value, max_value = get_min_max(List)
print(f'The minimum value in the list is: {min_value}')
print(f'The maximun value in the list is: {max_value}')
```
