# **Method No 1**

```python
sample = input("Enter a sentence: ")

sample_list = sample.split()
print(sample_list)


sample_list.pop(-1)
print(sample_list)

sample_list.insert(0, "World")
print(sample_list)

sample = " ".join(sample_list)
print(sample)
```

# **Method No 2**

```python
sentence = input("Enter the sentence: ")

new_sentence = sentence.split()
print(new_sentence)

new_sentence = new_sentence[::-1]
print(new_sentence)

new_sentence = " ".join(new_sentence)
print(new_sentence)
```

# **Method No 3**

```python
sentence = input("Enter the sentence: ")

new_sentence = sentence.split()
print(new_sentence)

new_sentence.reverse()
print(new_sentence)

reversed = " ".join(new_sentence)
print(reversed)
```
