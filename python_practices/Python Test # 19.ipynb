## **Count The Vowels**

```python
def count_vowels(text):
  vowels = 'aeiouAEIOU'
  return sum(1 for i in text if i in vowels)

user_input = input("Enter the text: ")
print('No of vowels:',count_vowels(user_input))
```

# **Check Word Is Palindrome Or Not**

```python
def is_palindrome(word):
    return word == word[::-1]

user_input = input("Enter a word: ")
print("Is palindrome:", is_palindrome(user_input))
```
