```python
def add(b):
  a = 2
  return a + b

sum = add(5)
print(sum)
```

```python
def add(b):
  a = 2
  c = a + b
  d = b - a
  return c, d

sum, sub = add(5)
print(sum)
print(sub)
```

```python
def disp(asd):
  def shows():
    return "Shows function "
  result = shows() + asd + " Disp function"
  return result

print(disp('Abdullah'))
```
